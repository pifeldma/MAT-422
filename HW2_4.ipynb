MLE for Random Samples

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from statsmodels import api
from scipy import stats
from scipy.optimize import minimize

x = np.linspace(-10,30,100)
e = np.random.normal(10,5,100)
y = 10 + 4*x + e
df = pd.DataFrame({'x':x,'y':y})
df.head()

sns.regplot(x='x',y='y',data=df)
plt.show()

features=api.add_constant(df.x)
model=api.OLS(y,features).fit()
model.summary()

res=model.resid
standard_dev=np.std(res)
standard_dev

def MLE_Norm(parameters): const, beta, std_dev = parameters
pred = 10 + 4*x
LL = np.sum(stats.norm.logpdf(y,pred,standard_dev))
neg_LL=-1*LL
```

Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x=np.array([5,10,15,20,25,30]).reshape((-1,1))
y=np.array([39,33,26,22,15,11])

model = LinearRegression().fit(x,y)

r_sq = model.score(x,y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept:{model.intercept_}")
print(f"slope: {model.coef_}")

```
