```python
import networkx as nx
import numpy as np

G = nx.random_regular_graph(3,10)
nx.draw(G)


A = nx.adjacency_matrix(G).toarray()
D = np.diag(np.ones(10) * 3)
L = D - A
L = nx.laplacian_matrix(G).toarray()
np.array_equal(nx.laplacian_matrix(G).toarray(), D - A)
```
