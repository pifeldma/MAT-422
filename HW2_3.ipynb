Independent Variables and Random Samples


Joint Probability Distribution

```python
import numpy as np
import seaborn as sns
import pandas as pd

A = np.random.normal(size=100)
B = np.random.normal(size=100)

df = pd.DataFrame({'A' : A , 'B' : B})
sns.jointplot(x='A', y = 'B' , data=df)
```

Correlation and Dependence

```python
import numpy as np
x=np.arange(10,20)
y=np.array([1,5,3,8,10,33,76,42,77,64])

r = np.corrcoef(x,y)
print(x)
print(y)
print(r)
```

Random Samples

```python
import random

list1 = [1,11,21,31,41]
print("List:", list1)
print(random.sample(list1, k=3))

place = ["Tucson", "Phoenix", "Flagstaff", "Yuma", "Sedona"]
print(random.sample(place, k=3))
```
